```python
import time
import pandas as pd
import tkinter as tk
from tkinter import filedialog

print("Elija su archivo excel")
# Crear una ventana oculta para evitar que aparezca la ventana de Tkinter
root = tk.Tk()
root.withdraw()

# Abrir el explorador de archivos y permitir que el usuario seleccione un archivo Excel
ruta_archivo = filedialog.askopenfilename(title="Selecciona un archivo Excel",
                                          filetypes=[("Archivos de Excel", "*.xlsx *.xls")])

# Verifica si el usuario seleccionó un archivo
if ruta_archivo:
    # Leer el archivo Excel
    df = pd.read_excel(ruta_archivo, engine='openpyxl')
    print("El archivo fue correctamente seleccionado")
    pass

else:
    print("No se seleccionó ningún archivo.")

print("Menú")
print("1. Bubblesort")
print("2. InsertionSort")
print("3. MegeSort")

opcion = int(input("Ingrese su eleccion: "))
if opcion ==1:
    def bubblesort(vector):
        print("El vector a ordenar es: ", vector)
        n = len(vector)  # obtener la longitud del vector
        for i in range(n-1):
            for j in range(0, n-i-1):
                if vector[j] > vector[j+1]:
                    vector[j], vector[j+1] = vector[j+1], vector[j]
        print("El vector ordenado es: ", vector)
        return vector

    for columna in df.columns:
        print(f"Ordenando la columna: {columna}")
        # Extraer la columna como lista
        vectorbs = df[columna].dropna().tolist()
        # Medir el tiempo de ejecución
        tiempo_inicial = time.time()
        vector_ordenado = bubblesort(vectorbs)  # Ordenar la columna
        tiempo_final = time.time()

        print(f"Tiempo de ejecución para la columna '{columna}': {tiempo_final - tiempo_inicial} segundos")
        print("\n")


elif opcion ==2:
    for columna in df.columns:
        arr = df[columna].dropna().tolist()
        def insertionSort(arr):
            for i in range(1,len(arr)):
                key = arr[i]
                j = i-1
                while j>=0 and key < arr[j]:
                    arr[j+1] = arr[j]
                    j-=1
                arr[j+1]= key
        def printArray(arr):
            print("Ordenando la columna")
            for i in range(len(arr)):
                print(arr[i], end =" ")

        tiempo = time.time()
        insertionSort(arr)
        printArray(arr)
        final_tiempo = time.time()
        print(f"Tiempo de ejecucion:{final_tiempo-tiempo} segundos")

elif opcion ==3:
    for columna in df.columns:
        arr = df[columna].dropna().tolist()
        def merge(arr, left, mid, right):
            n1 = mid - left + 1
            n2 = right - mid

            # Create temp arrays
            L = [0] * n1
            R = [0] * n2

            # Copy data to temp arrays L[] and R[]
            for i in range(n1):
                L[i] = arr[left + i]
            for j in range(n2):
                R[j] = arr[mid + 1 + j]

            i = 0  # Initial index of first subarray
            j = 0  # Initial index of second subarray
            k = left  # Initial index of merged subarray

            # Merge the temp arrays back
            # into arr[left..right]
            while i < n1 and j < n2:
                if L[i] <= R[j]:
                    arr[k] = L[i]
                    i += 1
                else:
                    arr[k] = R[j]
                    j += 1
                k += 1

            # Copy the remaining elements of L[],
            # if there are any
            while i < n1:
                arr[k] = L[i]
                i += 1
                k += 1

            # Copy the remaining elements of R[],
            # if there are any
            while j < n2:
                arr[k] = R[j]
                j += 1
                k += 1

        def merge_sort(arr, left, right):
            if left < right:
                mid = (left + right) // 2

                merge_sort(arr, left, mid)
                merge_sort(arr, mid + 1, right)
                merge(arr, left, mid, right)

        def print_list(arr):
            print("Ordenando la columna")
            for i in range(len(arr)):
                print(arr[i], end =" ")

        tiempo = time.time()
        merge_sort(arr, 0, len(arr) - 1)
        print_list(arr)
        final_tiempo = time.time()
        print(f"Tiempo de ejecucion:{final_tiempo-tiempo} segundos")
```
